```python
import pandas as pd 
```

```python
#load data, then drop the "unnamed" column , then rename the id of listings dataset so it can easily joined 

listings = pd.read_csv("listings.csv").drop('Unnamed: 0',axis=1).rename(columns={'id':'listing_id'})
neigh = pd.read_csv("neigh.csv").drop('Unnamed: 0',axis=1)
reviews = pd.read_csv("reviews.csv", parse_dates=True).drop('Unnamed: 0',axis=1)
```

```python
print(listings)
print(neigh)
print(reviews)
```

```python
data = reviews.merge(listings,on="listing_id")
```

```python
data = data.merge(neigh,on="neighbourhood")
```

```python
data.head()
```

```python
data.info()
```

```python
# Explore : group by neighbourhood 

data_by_neigh = data.groupby('neighbourhood').count()['name'].sort_values(ascending=False).reset_index()
data_by_neigh.rename(columns={'name':'count'},inplace=True)
data_by_neigh.head()
```

```python
# Explore : group by listing id 

data_by_listing_id = data.groupby("listing_id").count()[['name']].sort_values(ascending=False,by='name')
```

```python
data_by_listing_id.reset_index(inplace=True)
```

```python
data_by_listing_id.rename(columns={'name':'count'},inplace=True)
```

```python
data_fin_listing_id = data_by_listing_id.merge(listings,on="listing_id")
```

```python
data_fin_listing_id['d_index'] = data_fin_listing_id.index
data_fin_listing_id.head()
```

```python
# Explore : Group by room type ( just want to explore a bit )
data_by_type = data.groupby("room_type").count()['name'].sort_values(ascending=False).reset_index()
data_by_type.rename(columns={'name':'count'},inplace=True)
data_by_type
```

```python
data_by_date_n_type = data.groupby(["date","room_type"]).count()
data_by_date_n_type
```

```python
# Export top 50 listing ID to JSON , so can use it to display the data using javascript 
data_listing_json_top_50 = data_fin_listing_id.head(50).to_json("gb_listing.json",orient="records")
```
